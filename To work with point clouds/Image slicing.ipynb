```python
def split_file(file, win_size, hop, path, name):
    '''Функция разбиения изображения на меньшие изображения скользящим окном
    Param:
        file - входящий файл изображения
        win_size -  размер окна
        hop -  шаг смещения окна
        path - путь где сохраняться файлы
        name - имя под которым сохраняться файлы файлов
    '''
    n = 0
    for j in range(0, file.size[1] - win_size + 1, hop):
        for i in range(0, file.size[0] - win_size + 1, hop):
            box = (i, j, i+win_size, j+win_size)
            area = file.crop(box)
            filename = f'{path}{name}_{n}.tiff'  
            #filename = f'{path}{name}_{i}_{j}.tiff'                             # альтернативное именование файлов. переменная n будет не нужна
            area.save(filename)
            n += 1 
```

```python
win_size = 256                                                                  # размер окна
hop =100                                                                        # шаг смещения окна
path_x = '/content/'
path_y = '/content/'
name_x = 'ximg'
name_y = 'yimg'
```
